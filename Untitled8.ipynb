```python
# ==========================================
# IMPORT LIBRARIES
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# ==========================================
# LOAD DATASET
# ==========================================

from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns)

# ==========================================
# DATA CLEANING
# ==========================================

df.drop_duplicates(inplace=True)

# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# Convert order_time to datetime hour
df["hour"] = pd.to_datetime(
    df["order_time"],
    format="%H:%M"
).dt.hour

print("\nCleaning Completed")

# ==========================================
# Q1 PEAK HOUR DELAY PATTERN
# ==========================================

print("\n==============================")
print("Q1 - Peak Hour Delay Pattern")
print("==============================")

def peak_hour(hour):
    if (8 <= hour <= 10) or (17 <= hour <= 20):
        return "Peak"
    return "Off-Peak"

df["time_category"] = df["hour"].apply(peak_hour)

peak = df[df["time_category"] == "Peak"]["delay_mins"]
offpeak = df[df["time_category"] == "Off-Peak"]["delay_mins"]

print("Peak Average Delay:", round(peak.mean(),2))
print("Off-Peak Average Delay:", round(offpeak.mean(),2))

difference = peak.mean() - offpeak.mean()

print("Difference:", round(difference,2), "minutes")

t,p = ttest_ind(peak, offpeak)

print("P Value:", p)

if p < 0.05:
    print("Significant Difference Found")
else:
    print("No Significant Difference")

plt.figure(figsize=(8,5))
sns.boxplot(x="time_category",
            y="delay_mins",
            data=df)
plt.title("Peak vs Off Peak Delay")
plt.show()

# ==========================================
# Q2 WEATHER VS DELAY
# ==========================================

print("\n==============================")
print("Q2 - Weather vs Delay")
print("==============================")

weather_delay = df.groupby(
    "weather_condition"
)["delay_mins"].median()

print(weather_delay)

plt.figure(figsize=(8,5))
sns.boxplot(
    x="weather_condition",
    y="delay_mins",
    data=df
)
plt.title("Weather vs Delay")
plt.show()

rain_df = df[
    df["weather_condition"] == "Rain"
]

rain_impact = rain_df.groupby(
    "order_type"
)["delay_mins"].median()

print("\nRain Impact by Order Type")
print(rain_impact.sort_values(ascending=False))

plt.figure(figsize=(8,5))
rain_impact.sort_values().plot(kind="bar")
plt.title("Order Type Impact During Rain")
plt.ylabel("Median Delay")
plt.show()

# ==========================================
# Q3 RIDER EXPERIENCE EFFECT
# ==========================================

print("\n==============================")
print("Q3 - Rider Experience Effect")
print("==============================")

less_exp = df[
    df["rider_experience_yrs"] < 2
]["delay_mins"]

more_exp = df[
    df["rider_experience_yrs"] > 4
]["delay_mins"]

print("Average Delay (<2 years):",
      round(less_exp.mean(),2))

print("Average Delay (>4 years):",
      round(more_exp.mean(),2))

t,p = ttest_ind(
    less_exp,
    more_exp
)

print("P Value:", p)

if p < 0.05:
    print("Experience Significantly Affects Delay")
else:
    print("No Significant Difference")

# ==========================================
# Q4 DASHBOARD DATA
# ==========================================

print("\n==============================")
print("Q4 - Dashboard Metrics")
print("==============================")

df["on_time"] = np.where(
    df["delay_mins"] <= 10,
    1,
    0
)

city_rate = (
    df.groupby("city")["on_time"]
    .mean() * 100
)

monthly_delay = (
    df.groupby("order_date")["delay_mins"]
    .mean()
)

vehicle_delay = (
    df.groupby("vehicle_type")["delay_mins"]
    .mean()
)

print("\nCity Wise On Time Rate")
print(city_rate)

# Panel 1
plt.figure(figsize=(10,5))
city_rate.sort_values().plot(
    kind="bar"
)
plt.title("City Wise On Time Rate")
plt.ylabel("Percentage")
plt.show()

# Panel 2
plt.figure(figsize=(10,5))
vehicle_delay.sort_values().plot(
    kind="bar"
)
plt.title("Vehicle Type Comparison")
plt.ylabel("Average Delay")
plt.show()

# Monthly Trend
df["order_date"] = pd.to_datetime(
    df["order_date"]
)

monthly = (
    df.groupby(
        df["order_date"].dt.month
    )["delay_mins"]
    .mean()
)

plt.figure(figsize=(10,5))
monthly.plot(marker="o")
plt.title("Monthly Delay Trend")
plt.xlabel("Month")
plt.ylabel("Average Delay")
plt.grid(True)
plt.show()

print("\nAnalysis Completed Successfully")
```
